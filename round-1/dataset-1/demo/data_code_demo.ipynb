# Sentence-Level Readability Dataset DemoThis notebook demonstrates how to load and explore the sentence-level readability datasets (WeeBIT and CEFR-SP) for the paper "The Uniformity Principle: How Within-Sentence Variability Predicts Readability".## What this artifact doesThis dataset artifact provides:- **WeeBIT dataset**: 3,125 sentences with 5 age intervals from Vajjala & Meurers (2012)- **CEFR-SP dataset**: 10,004 sentences annotated with CEFR levels (A1-C2) by English education professionals (EMNLP 2022)The datasets have been standardized to a common schema with:- `input`: sentence text- `output`: readability score as string (normalized to 0-1 scale, 0=easy, 1=hard)- `metadata_*`: additional metadata fields## Demo OverviewIn this notebook, we will:1. Load a small demo dataset2. Explore the data structure3. Display basic statistics about the dataset

## Basic Dataset StatisticsCalculate basic statistics about the demo dataset.

```python
# Compute basic statisticsdataset = data["datasets"][0]examples = dataset["examples"]# Extract readability scores (as floats)scores = [float(ex["metadata_readability_score"]) for ex in examples]print(f"Dataset: {dataset['dataset']}")print(f"Number of examples: {len(examples)}")print(f"\nReadability Score Statistics:")print(f"  Min: {min(scores)}")print(f"  Max: {max(scores)}")print(f"  Mean: {mean(scores):.4f}")if len(scores) > 1:    print(f"  Std Dev: {stdev(scores):.4f}")# Count by age intervalage_intervals = Counter(ex.get("metadata_complexity_age_interval") for ex in examples)print(f"\nExamples by Age Interval:")for interval, count in sorted(age_intervals.items()):    print(f"  Interval {interval}: {count} examples")# Text length statisticstext_lengths = [len(ex["input"].split()) for ex in examples]print(f"\nText Length (words) Statistics:")print(f"  Min: {min(text_lengths)}")print(f"  Max: {max(text_lengths)}")print(f"  Mean: {mean(text_lengths):.2f}")
```
